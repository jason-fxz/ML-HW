```python
import pandas as pd
from statsmodels.tsa.seasonal import STL

# ,HUFL,HULL,MUFL,MULL,LUFL,LULL,OT
# 1.1 读取 ETTh1 OT 序列，假设 ts 是一个 pandas.Series，index 为 DatetimeIndex
ts_raw = pd.read_csv('../datasets/ETTh1.csv', parse_dates=['date'], index_col='date')['OT']
```

```python
in_len = 720 * 12 * 2
out_len = 720 * 4
s = 0
ts = ts_raw[s : s + in_len]
ts_next = ts_raw[s + in_len : s + in_len + out_len]

stl = STL(ts, period=24 * 4)      # period=24 表示一天的周期（小时级数据）
res = stl.fit()
trend = res.trend              # 长期趋势
seasonal = res.seasonal        # 周期成分
residual = res.resid           # 残差


# 可视化

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.subplot(4, 1, 1)
plt.plot(ts, label='Original')
plt.plot(ts_next, label='Next', color='red', alpha=0.7)
plt.legend(loc='upper left')
plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(4, 1, 4)
plt.plot(residual, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
